```python
import pickle
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
with open('/content/drive/MyDrive/LSTM data/only_nlp_tech_available_tokens_train_16.5.pkl','rb') as fi:
  all_lst = pickle.load(fi)
```

```python
df = pd.concat(all_lst)
```

```python
df.info()
```

```python
features_list = ['macd', 'macdsignal', 'macdhist', 'rsi', 'ema', 'adx',
       'aroondown', 'aroonup', 'adosc', 'upperband', 'middleband', 'lowerband',
       'real', 'obv', 'slowk', 'slowd', 'macd_f1', 'macd_f2', 'macd_f3',
       'adosc_f1', 'adosc_f2', 'adosc_f3', 'rsi_f1', 'rsi_f2', 'ema_f2',
       'ema_f1', 'scoh_f1', 'scoh_f2', 'scoh_f3', 'scoh_f4', 'adx_f1',
       'adx_f2', 'aroon_f1', 'aroon_f2', 'bband_f1', 'bband_f2','num_tweets',
       'num_positive_tweets', 'num_negative_tweets', 'num_neutral_tweets',
       'NUM_POSITIVE_LIKES', 'NUM_NEGATIVE_LIKES', 'NUM_NEUTRAL_LIKES',
       'daily_sentiment', 'daily_sentiment_30_days', 'long_term_sentiment',
       'relative_tweet_volume', 'num_unique_accounts']
```

```python
from scipy.stats import chi2_contingency
```

```python
dt = {'feature':[],'p':[]}
for ft in features_list:
  stat, p, dof, expected = chi2_contingency(pd.crosstab(df[ft], df['target']))
  dt['feature'].append(ft)
  dt['p'].append(p)



```

```python
pd.DataFrame(dt)
```

```python
pd.DataFrame(dt).to_csv('Chi_Square_16_5.csv',index=False)
```
